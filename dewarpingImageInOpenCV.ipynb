```python
import cv2
import matplotlib.pyplot as plt
from google.colab.patches import cv2_imshow
print(cv2.__version__)
import numpy as np

img = cv2.imread('warp_img.jpg')

# Pixel values in original image
top_left = [40, 70]
top_right = [576, 10]
bottom_left = [3, 490]
bottom_right = [600, 522]

# Create point matrix
point_matrix = np.float32([top_left, top_right, bottom_left, bottom_right])

# Draw circle for each point
for x in range(0,4):
  cv2.circle(img, (point_matrix[x][0], point_matrix[x][1]), 10, (0, 0, 255), cv2.FILLED)

# Output image size
width, height = 250, 350
```

```python
# Desired points value in output images
converted_top_left_pixel_value = [0, 0]
converted_top_right_pixel_value = [width, 0]
converted_bottom_left_pixel_value = [0, height]
converted_bottom_right_pixel_value = [width, height]

# Convert points
converted_points = np.float32([converted_top_left_pixel_value, converted_top_right_pixel_value,
                               converted_bottom_left_pixel_value, converted_bottom_right_pixel_value])

```

```python
# perspective transform
perspective_transform = cv2.getPerspectiveTransform(point_matrix, converted_points)
img_Output = cv2.warpPerspective(img, perspective_transform, (width, height))
```
